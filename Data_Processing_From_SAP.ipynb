```python
#specify the path to location of your excel file
data = pd.read_csv(r'C:\Users\dtu\Desktop\SAP Data\New folder\Row_A.csv', header = 2, low_memory = False)
df1 = pd.DataFrame(data)
#display (df1)
df1.loc[0:2, :]
```
